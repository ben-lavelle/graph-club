## A Max Flow / Min Cut problem in the US Postal Service, Georgia

It's mid-2020. You are celebrated voting rights activist [Stacey Abrams](https://en.wikipedia.org/wiki/Stacey_Abrams), and have been fighting voter suppression across Georgia for years. This year's election is looking a little different, and postal voting will be more important than ever across many different communities.

You're concerned with the ability of the US Postal Service to carry the expected volume of mail-in ballots from historic centres like [Savannah](https://en.wikipedia.org/wiki/Savannah,_Georgia) on the state's East Coast to the capital, [Atlanta](https://en.wikipedia.org/wiki/Atlanta), for counting* and want to make sure there's enough capacity to get all the ballots to the count in time.

(* *This ~probably~ definitely isn't how it works but roll with it*).

You know postal votes are carried by road from town to town, in postal vans with different capacities suited to the variety of roads found in Georgia, from large interstates to small country tracks. Some short suburban routes can even take several trips a day. Most routes can transfer ballots at a rate of between 100 and 1000 votes per day.

![mock of georgia road network](img/georgia_network_map_unweighted.png)

You've done your surveys and collected the capacity data for each route [see imports below from `usps_georgia.py`]. You want to know:

1. **What's the maximum number of votes that can flow from Savannah ('S') to Atlanta ('T') per day?** Assume a steady state solution.
2. If you have a limited crowd-funded budget that can, say, add a total of 500 capacity to the network spread over as many routes as needed, **on which routes should you add capacity to best improve the capacity of the network overall?**

```python
import networkx as nx
from usps_georgia import nodes, ud_edges_nx
```

```python
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(ud_edges_nx)
```

```python
import matplotlib.pyplot as plt
color_map = []
for node in G.nodes:
    if node in ['S', 'T']:
        color_map.append('red')
    else:
        color_map.append('cyan')
    
nx.draw_networkx(G, node_color=color_map, with_labels=True)
```
