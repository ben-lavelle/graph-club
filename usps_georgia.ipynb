```python
from usps_georgia import nodes, ud_edges_nx
import networkx as nx
```

```python
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(ud_edges_nx)
```

```python
import matplotlib.pyplot as plt
color_map = []
for node in G.nodes:
    if node in ['S', 'T']:
        color_map.append('red')
    else:
        color_map.append('cyan')
    
nx.draw_networkx(G, node_color=color_map, with_labels=True)
```

```python
nx.maximum_flow_value(G, _s='S', _t='T', capacity='capacity')
nx.maximum_flow(G, _s='S', _t='T', capacity='capacity')
```

```python
mc = nx.minimum_cut(G, _s='S', _t='T', capacity='capacity')
mc
```

```python
color_map = []
for node in G.nodes:
    if node in ['S', 'T']:
        color_map.append('red')
    elif node in mc[1][0]:
        color_map.append('green')
    else:
        color_map.append('cyan')
    
nx.draw_networkx(G, node_color=color_map, with_labels=True)
```
